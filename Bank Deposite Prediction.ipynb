```python
import graphlab as gl
```

```python
data = gl.SFrame('data/bank-train-data.csv')
```

```python
data
```

```python
train_data, test_data = data.random_split(.8, seed =0)
```

```python
model =  gl.logistic_classifier.create(train_data, target='y', features=['job','marital','housing' ,'balance' , 'loan'])
```

```python
predictions = model.classify(test_data)
```

```python
results = model.evaluate(test_data)
```
